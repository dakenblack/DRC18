# Naive Bayes Classifier

```python
import numpy as np
import matplotlib.pyplot as plt
# Note matplotlib needs Pillow installed to import jpg and bmp images
import matplotlib.image as mi
%matplotlib inline
```

```python
def train(filel, fileu):
    # @Param file1 string with name of labelled image
    # @Param fileu string with name of unlabelled image
    
    # Read in labelled and unlabelled image, and reshape so that the matrices represent a list of RGB tuples
    lab = mi.imread(filel)
    unl = mi.imread(fileu)
    lab = np.reshape(lab, (-1, 3))
    unl = np.reshape(unl, (-1, 3))
    
    # find mask of pixels that represent the painted elements in the labelled image
    b = np.logical_and(np.logical_and(lab[:,0] == 255, lab[:,1] == 0) , lab[:,2] == 255) 
    # find on (painted) and off (not painted) pixels
    on = unl[b,:]
    off = unl[~b,:]
    
    # initialize the classifier output
    ret = np.zeros((256*256*256,1))
    # for every on pixel increment the rgb index in the classifier
    for clr in on:
        ret[clr[2]*256*256 + clr[1]*256 + clr[0]] += 1;
    # for every on pixel decrement the rgb index in the classifier
    for clr in off:
        ret[clr[2]*256*256 + clr[1]*256 + clr[0]] -= 1;
        
    mn = ret.min();
    mx = ret.max();
    ret = (ret - mn) / (mx - mn)
    return ret;

def classify(I, cl, t):
    # @Param I numpy array of image to be classified (See cv2, Pillow, or Matplotlib.image)
    # @Param cl classifier output from train function
    # @Param t threshold
    I = I.astype(int)
    ids = I[:,:,2]*256*256 + I[:,:,1]*256 + I[:,:,0]
    scores = cl[ids];
    scores = np.reshape(scores, (scores.shape[0], scores.shape[1]))
    mask = scores > t;
    mask = mask.astype(int);
    return mask, scores;
```

```python
cl = train('labelled1.bmp','unlabelled1.jpg');
```

```python
fileu = 'unlabelled2.jpg'
I = mi.imread(fileu)
# Tweak the threshold value as you see fit
(m,_) = classify(I, cl, 0.837)
```

```python
plt.subplot(2,1,1);
plt.imshow(I)
plt.subplot(2,1,2);
plt.imshow(m)
```

## Explaination

* The classifier output is a list of 1677216(256 x 256 x 256) values. Each element is associated to an RGB value,
    * $index\ in\ classifier\ list = b \times 256 \times 256 + g \times 256 + r$ 
    * where r,g,b are the intensities of the red, blue and green channels
* Each element holds the score of that RGB color code being the activated pixel
    * See the plot of the classifier list below
    * The scores go from 0 to 1.0
